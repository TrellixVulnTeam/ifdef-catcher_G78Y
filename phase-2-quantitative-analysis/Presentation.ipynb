# Part 1: Analyze 33 projects using cppstats

## Definitions

- version 0: old version of project to be analyzed. Is 2008/2009 version, or something close to that.
- version 1: new version of project to be analyzed. Is 2020/2021 version, or something close to that.
- metric diff: differentiating a metric between version 0 and 1. E.g., discipline diff.

## Structure and Algorithms

This project use [cppstats](https://github.com/clhunsen/cppstats), and the cppstats project use [srcML](https://www.srcml.org/). The idea is simple: for an target project, transform all files in xml using srcML, and get the ifdef tags. After that, analyze the structure of this collected data.

For each project, download all the code to a local directory and run cppstats. The cppstats tool will generate a csv file with the metrics.

## Threat

From version 0 to version 1, there is certainly unchange code. So, to investigate if there are changes to disciplined or undisciplined code, this immutable files is harmful for analysis, as it can change the results.

# Part 2: Remove duplicates

## Idea

Remove equal files in both versions. And files with no changes on #ifdef blocks. With this, we can remove the threat from Part 1.

## Structure and Algorithms

The cppstats have 2 phases: preparation and analysis.

- Preparation: transform each file of each project to xml, and do some treatment.
- Analysis: find ifdefs blocks and count disciplined/undisciplined code.

After preparation and before analysis, we can remove equals files and files with equals ifdef blocks. With this, is possible to analyze only the changes between version 0 and 1.

So, the operation between preparation and analysis we call *filter*. The filter operation do the following tasks:

- Remove equal files
- Remove Files with equal ifdef blocks. Not any block, but all blocks on each file.
